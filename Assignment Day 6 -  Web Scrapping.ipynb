Parsing kedalam HTML

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

Mengambil datanya

```python
#Apabila menggunakan select, maka akan mengambil semua datanya dalam bentuk array
list_content = soup.select('.cds-ProductCard-gridCard')
```

Looping dan parsing sesuai data yang kita butuhkan

```python
#Apabila menggunakan select one, hanya mengambil satu data saja dan dalam bentuk string maka bisa diubah menjadi text
#Limit 1 agar gampang ngeceknya
for item in list_content[:4]:
  data = dict()

  #data collection
  data['Title'] = item.select_one('.cds-CommonCard-title').text
  data['Keterampilan'] = ', '.join(item.select_one('b.css-l47nsn').next_sibling.strip().split(', ')[:3])
  data['Rating'] = item.select_one('span.css-6ecy9b').text.replace('.', ',')
  data['Ulasan'] = str(int(float(item.select_one('div.css-vac8rf').text.replace(' reviews', '').replace('K', '').strip()) * (1000 if 'K' in item.select_one('div.css-vac8rf').text else 1)))
  data['Tingkat'] = item.select_one('.cds-CommonCard-metadata').text.split('·')[0].strip().replace('Â', '').rstrip()
  data['Produk'] = item.select_one('.cds-CommonCard-metadata').text.split('·')[1].strip().replace('Â', '').rstrip()
  data['Durasi'] = item.select_one('.cds-CommonCard-metadata').text.split('·')[2].strip().replace('Â', '').rstrip()
  print(data)
```
